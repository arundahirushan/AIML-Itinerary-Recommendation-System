```python
import pandas as pd

# load your cleaned dataset
df = pd.read_csv("data_with_clusterID.csv")

# compute average distance per cluster
cluster_distance = df.groupby('Cluster_Name')['real_distance'] \
                     .mean() \
                     .reset_index() \
                     .rename(columns={'real_distance': 'Avg_Distance'})

print(cluster_distance)
cluster_distance.to_csv("cluster_distance.csv", index=False)
```

```python
cluster_features = pd.read_csv("cluster_features.csv")

cluster_features = cluster_features.merge(cluster_distance, 
                                          on='Cluster_Name', 
                                          how='left')

cluster_features.to_csv("cluster_with_distance.csv", index=False)
```

```python
df = cluster_features
df = df.drop(columns=['Hilly', 'Off-Road', 'Paved'], errors='ignore')
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Avg_Distance']] = scaler.fit_transform(df[['Avg_Distance']])
df[['Num_Places']] = scaler.fit_transform(df[['Num_Places']])

```

```python
print(df.head())
df.to_csv("final_cluster_features.csv", index=False)
```
